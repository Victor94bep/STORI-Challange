```python
## Question 2


## Creating fuction to remove text in column  cust_id

def dejar_solo_cifras(txt):
  return "".join(c for c in txt if c.isdigit())

## Applying fucntion to column  cust_id 
## Excluding letters

df.cust_id = df.cust_id.map(dejar_solo_cifras)

## Changing date format: activated_date [yyyy/mm] & last_payment_date [yyyy/mm/dd]
## Creating and Adding: activated_date1 & last_payment_date1 

df['activated_date'] = pd.to_datetime(df.activated_date)
df['activated_date1'] = df['activated_date'].dt.strftime('%Y/%m')
df['last_payment_date'] = pd.to_datetime(df.last_payment_date)
df['last_payment_date1'] = df['last_payment_date'].dt.strftime('%Y/%m/%d')



## Setting fuction to create new Colum  percentage

def percentage (fila):
    result=(fila['cash_advance'] / fila['credit_limit']) *100
    return result 
## Applying percentage function and crating new dataframe column

df['PERCENTAGE']=df.apply(percentage, axis=1)


## Creating new dataframe filtered by condition of time  in last_payment_date & activated_date

filtered_df =df.loc[(df["last_payment_date"].between('2020-01-1', '2020-12-31')) & (df["activated_date"].between('2020-01-31', '2020-12-31'))]

 ## Using iloc to show filtered_df 
 ## Reporting dataframe filtered 

filtered_df.loc[:,['cust_id','activated_date', 'last_payment_date','cash_advance','credit_limit','PERCENTAGE']]
```
